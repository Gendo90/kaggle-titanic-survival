```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
train_df = pd.read_csv("train.csv")
train_df = train_df.dropna(axis="columns", how="all")
train_df
```

```python
train_df.drop(["Cabin", "Ticket", "Name", "PassengerId"], axis=1, inplace=True)
train_df
```

```python
boolean_df = train_df.isna()

for i in range(boolean_df.index.size):
    for j in range(8):
        if(boolean_df.iloc[i, j]):
            print("row: {} column: {}".format(i, boolean_df.columns[j]))
```

```python
# ignore Age and Embarked for now - set later based on Name, ticket price, etc.
# ignore Embarked as well! Need to make multiple columns for the three values
train_df.drop(["Fare", "Age", "Embarked"], axis=1, inplace=True)

# train_df = train_df.dropna()
train_df
```

```python
# convert non-numeric data to numbers
train_df["Sex"] = train_df["Sex"].apply(lambda x: 0 if(x=="male") else 1)
train_df
```

```python

train_df["Survived"].value_counts()
```

## Create a Train/Test Split

```python
selected_features = train_df[["Pclass", "Sex", "SibSp", "Parch"]]
y = train_df["Survived"]
y.array
```

```python
from sklearn.model_selection import train_test_split

X = selected_features

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, train_size=0.8, random_state=42)
```

```python
X_train.head()
```

```python
y_train
```

## Pre-processing (Normalize data)

```python
# Scale your data

#setup MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

#set initial sizing of minmaxscaler using training data
X_scaler = MinMaxScaler().fit(X_train)
```

```python
#use MinMaxScaler to transform the data in the training and test data sets
#so that it is not weighted improperly

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Train the Model

```python
classifier = LogisticRegression()
classifier
```

```python
#create the model using the training data!
classifier.fit(X_train_scaled, y_train["Survived"].array)
```

## Hyperparameter Tuning

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [1, 5, 10, 50, 100, 200, 500],
              'max_iter': [100, 200, 300],
              'tol': [.0001, 0.001, 0.005]}

grid = GridSearchCV(classifier, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
#compare with the test data!
predictions = grid.predict(X_test_scaled)
```

```python
#check classification report for how the tuned model did!
from sklearn.metrics import classification_report

print(classification_report(y_test, predictions,
                            target_names=["Actual", "Predicted"]))
```
