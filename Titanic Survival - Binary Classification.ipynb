## Create a Train/Test Split

```python
selected_features = train_df[["Pclass", "Sex", "SibSp", "Parch", "Fare"]]
y = train_df["Survived"]
y.array
```

```python
from sklearn.model_selection import train_test_split

X = selected_features

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, train_size=0.75, random_state=31)
```

```python
X_train.head()
```

```python
y_train
```

## Pre-processing (Normalize data)

```python
# Scale your data

#setup MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

#set initial sizing of minmaxscaler using training data
X_scaler = MinMaxScaler().fit(X_train)
```

```python
#use MinMaxScaler to transform the data in the training and test data sets
#so that it is not weighted improperly

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Train the Model

```python
classifier = LogisticRegression()
classifier
```

```python
#create the model using the training data!
classifier.fit(X_train_scaled, y_train.array)
```

## Hyperparameter Tuning

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [1, 5, 10, 50, 100, 200, 500],
              'max_iter': [100, 200, 300],
              'tol': [.0001, 0.001, 0.005]}

grid = GridSearchCV(classifier, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
#compare with the test data!
predictions = grid.predict(X_test_scaled)
```

```python
#check classification report for how the tuned model did!
from sklearn.metrics import classification_report

print(classification_report(y_test, predictions,
                            target_names=["Actual", "Predicted"]))
```

```python
y_test
```

## Setup Random Forest Classification

```python
#uses a standard random forest for this notebook
from sklearn.ensemble import RandomForestClassifier

random_forest_classifier = RandomForestClassifier(n_estimators=200, class_weight="balanced")
random_forest_classifier
```

```python
#create the model using the training data!
random_forest_classifier.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {random_forest_classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {random_forest_classifier.score(X_test_scaled, y_test)}")
```

## Set Random Forest Hyperparameter Tuning

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV

param_grid = {'ccp_alpha': [0, 0.5, 1],
             'n_estimators': [100, 200, 300],
              'class_weight': ["balanced",]
             }

grid2 = GridSearchCV(random_forest_classifier, param_grid)
```

```python
# Train the model with GridSearch
grid2.fit(X_train_scaled, y_train)
```

```python
predictions_random_forest = grid2.predict(X_test_scaled)

print(classification_report(y_test, predictions_random_forest,
                            target_names=["Actual", "Predicted"]))
```
